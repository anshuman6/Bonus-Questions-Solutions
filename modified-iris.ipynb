```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("modified_iris.csv")
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,7.5))
plt.scatter(df["sepal_length"], df["sepal_width"], c=df["petal_length"], alpha = 0.5, s=df["petal_width"], cmap='RdBu')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.axis([4,8,2,4.5])
plt.clim(min(df.petal_length), max(df.petal_length))
plt.colorbar(orientation='vertical', label='petal Length')
plt.show()
```
